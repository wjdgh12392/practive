```python
data_set = np.random.rand(20, 3)
data_set
```

```python
df = pd.DataFrame(data_set, columns=['A', 'B', 'C'])
df
```

```python
df.plot()
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'barh')
```
